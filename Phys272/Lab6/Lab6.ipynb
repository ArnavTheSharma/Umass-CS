```python
theta1 = [0.34, 0.5, 0.64, 0.76]
theta2 = [0.25, 0.35, 0.44, 0.52]
```

```python
import numpy as np
import matplotlib.pyplot as plt

coefficients = np.polyfit(theta1,theta2, 1)
print(f"Polynomial coefficients: {coefficients}")

# Create a polynomial function from the coefficients
poly_function = np.poly1d(coefficients)

# Generate points for plotting the fitted curve
x_new = np.linspace(0, 1, 100)
y_fitted = poly_function(x_new)
    
# Plot the original data and the fitted curve
plt.scatter(theta1,theta2, label='Original Data')
plt.plot(x_new, y_fitted, color='red', label='Fitted Polynomial')
plt.xlabel('sin(theta1)')
plt.ylabel('sin(theta2)')
plt.title('Polynomial Fit with numpy.polyfit')
plt.legend()
plt.grid(True)
plt.show()
```

```python
d0 = np.array([10.9, 89.6, 69, 11.2, 12.1, 48, 22.7, 16])
di = np.array([89.1, 10.4, 11, 68.8, 47.9, 12, 17.3, 24])



coefficients = np.polyfit(1/d0,1/di, 1)
print(f"Polynomial coefficients: {coefficients}")

# Create a polynomial function from the coefficients
poly_function = np.poly1d(coefficients)

# Generate points for plotting the fitted curve
x_new = np.linspace(0, 0.12, 100)
y_fitted = poly_function(x_new)

# Plot the original data and the fitted curve
plt.scatter(1/d0,1/di, label='Original Data')
plt.plot(x_new, y_fitted, color='red', label='Fitted Polynomial')
plt.xlabel('1/d0')
plt.ylabel('1/di')
plt.title('Polynomial Fit with numpy.polyfit')
plt.legend()
plt.grid(True)
plt.show()
```
